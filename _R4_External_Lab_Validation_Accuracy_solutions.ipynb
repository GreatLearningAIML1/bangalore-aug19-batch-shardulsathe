### Drop the original variables which are converted to dummy

Since 'foreign_worker' ,'telephone' columns have binary values, one of the dummy columns for each of these will be redundant.

```python
credit_dummies_df.drop(columns=['foreign_worker_no','telephone_none'],inplace=True)
```

```python
credit_dummies_df.corr().head()
```

```python
#Check which columns are highly correlated
corr_matrix = credit_dummies_df.corr().abs()
threshold = 0.8
# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than threshold
high_corr_cols = [column for column in upper.columns if any(upper[column] > threshold)]
high_corr_cols
```

### Q3 Split Train/Test data 70:30 ratio( 1 Marks)
##### Hint:from sklearn.model_selection import train_test_split

```python
array = credit_dummies_df.values
y = array[:,:1] # First column is dependent variable
X = array[:,1:]
```

```python

test_size = 0.30 # taking 70:30 training and test set
seed = 7  # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

### Q4 Build Random Forest Model( 1 Marks)
#### Hint:from sklearn.ensemble import RandomForestClassifier using n_jobs=2,n_estimators=500,criterion="entropy",random_state=9999

```python
model = RandomForestClassifier(n_jobs=2,n_estimators=500,criterion="entropy",random_state=9999)
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
model_score = model.score(X_test, y_test)
model_score_f1 = f1_score(y_test,y_predict)
accuracy = accuracy_score(y_test,y_predict)
print('model score: {}'.format(model_score))
print('f1 score: {}'.format(model_score_f1))
print('accuracy: {}'.format(accuracy))
```

### Q5 Calculate Confusion Matrix and Accuracy score (1 Marks)
##### Hint: Use confusion_matrix and accuracy_score

```python
print(metrics.classification_report(y_test, y_predict))
print(metrics.confusion_matrix(y_test, y_predict))
```
