```
bank_df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True) #Drop columns that are not required 
```

```
bank_df.shape
```

```
#Convert categorical String cols to numeric
le = preprocessing.LabelEncoder()
bank_df['Geography'] = le.fit_transform(bank_df['Geography'])
```

```
bank_df['Gender'] = le.fit_transform(bank_df['Gender'])
```

```
bank_df.info()
```

```
bank_df.head()
```

```
cols = bank_df.columns
```

```
cols[:-1]
```

```
#Distinguish Feature and Target set
X = bank_df[cols[:-1]]
y = bank_df['Exited']
```

```
X.head()
```

```
y.head()
```

```
#Convert features and labels to numpy array
X = X.values
y=y.values
X = X.astype('float32')
y = y.astype('int32')
```

```
X
```

```
y
```

```
#Split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```
#Normalize features
X_train  = tf.math.l2_normalize(X_train,axis=None,epsilon=1e-12,name=None)
X_test  = tf.math.l2_normalize(X_test,axis=None,epsilon=1e-12,name=None)
```

```
X_train
```

```
# Initialize the constructor
model = tf.keras.Sequential()
```

```
# Add input layers 
total_features = len(cols[:-1])
model.add(tf.keras.layers.Dense(total_features+1, activation='relu', input_shape=(total_features,)))
model.add(tf.keras.layers.Dense(total_features+1, activation='relu'))
#model.add(tf.keras.layers.Dense(total_features, activation='relu'))
#model.add(tf.keras.layers.Dense(total_features, activation='relu'))

```

```
#Output layer
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```
adam  = tf.keras.optimizers.Adam(learning_rate=0.03)
model.compile(loss='binary_crossentropy',
              optimizer='adam',
          metrics=['accuracy'])
```

```
history = model.fit(X_train, y_train,epochs=5, batch_size=1, verbose=1)
```

```
y_pred = np.round(model.predict(X_test))
score = model.evaluate(X_test, y_test,verbose=1)

print(score)
```

```
metrics.confusion_matrix(y_test, y_pred)
```
