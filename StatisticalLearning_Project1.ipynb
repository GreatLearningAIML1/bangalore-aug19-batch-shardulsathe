**Q3.** Print the list of teams which came Top 5 in terms of points

**Q4.** Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count(df):
    df = df.astype({"GoalsFor": int,"GoalsAgainst": int}) #change datatype from string to integer
    df['Goal_diff_count'] = df['GoalsFor'] - df['GoalsAgainst']
    highest = df[df['Goal_diff_count'] == df['Goal_diff_count'].max()]['Team']
    lowest = df[df['Goal_diff_count'] == df['Goal_diff_count'].min()]['Team']
    return df,highest,lowest
```

```python
laliga,highest,lowest = Goal_diff_count(laliga)
print("{} team has the highest goal difference whereas {} has the lowest goal difference.".format(highest.item(),lowest.item()))
```

**Q5.:** Create a new column with name “Winning Percent” and append it to the data set.
Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentag

```python
laliga = laliga.astype({"GamesWon": int,"GamesPlayed": int}) #change datatype from string to integer
laliga['GamesPlayed'].describe()
```

```python
#Since some of the teams haven't played any games, we need to handle 'divide by zero' exception
laliga['Winning Percent'] = (100.0 * laliga['GamesWon'])/laliga['GamesPlayed']
laliga['Winning Percent'] = laliga['Winning Percent'].fillna(0)
```

```python
laliga.head()
```

```python
laliga[laliga['Winning Percent'] == laliga['Winning Percent'].min()]
```

```python
laliga[laliga['GamesPlayed'] == laliga['GamesPlayed'].min()]
```

```python
laliga = laliga.sort_values('Winning Percent', ascending=False)
laliga['Team'].head() #Top 5 teams with highest winning percentage
```
